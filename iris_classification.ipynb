## Train-Test split

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.25)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
X_train = df_train.drop(columns=["target", "target_name"]).values
Y_train = df_train["target"].values
```

```python
X_train.shape
```

```python
Y_train
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

### Using validation set to evaluate the model

```python
# Splitting off a validation set from the training data
Xt, Xv, Yt, Yv = train_test_split(X_train, Y_train, test_size=0.25)
```

```python
model.fit(Xt, Yt)
```

```python
y_pred = model.predict(Xv)
```

```python
np.mean(y_pred == Yv)
```

```python
model.score(Xv, Yv)
```

### Using cross-validation to evaluate the model

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
```

```python
model2 = LogisticRegression()
```

```python
accuracies = cross_val_score(model2, X_train, Y_train, cv=5, scoring="accuracy")
```

```python
np.mean(accuracies)
```

```python
cross_preds = cross_val_predict(model, X_train, Y_train, cv=5)
```

```python
predicted_correctly_mask = cross_preds == Y_train
```

```python
X_train[~predicted_correctly_mask]
```

```python
df_predictions = df_train.copy()
```

```python
df_predictions["correct_prediction"] = predicted_correctly_mask
```

```python
df_predictions["prediction"] = cross_preds
```

```python
df_predictions["prediction_label"] = df_predictions["prediction"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
```

```python
df_predictions.head()
```

```python
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="prediction_label", data=df_predictions)
```

```python
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="target_name", data=df_predictions)
```

```python
def plot_incorrect_predictions(df_pred, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="prediction_label", data=df_pred, ax=axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_name", data=df_pred, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="correct_prediction", data=df_pred, ax=axs[2])
    axs[3].set_visible(False)

    plt.show()
```

```python
plot_incorrect_predictions(df_predictions, "petal length (cm)", "petal width (cm)")
```

```python
plot_incorrect_predictions(df_predictions, "sepal length (cm)", "sepal width (cm)")
```

We can see from the plots above that the first set of plots had the two wrong predictions overlapping. That is why we could only see one wrong prediction point in the graphs. If we plot by sepal length and width we can identify the two wrong predictions clearly. I also printed the dataframe rows of the wrong predictions, just to make sure of their number.

```python
print(df_predictions[df_predictions["correct_prediction"] == False].to_string())
```

## Model Tuning

```python
for reg_param in (1, 1.3, 1.8, 1.9, 2, 2.1, 2.3, 2.5):
    print(reg_param)
    lmodel = LogisticRegression(max_iter=200, C=reg_param)
    laccuracies = cross_val_score(lmodel, X_train, Y_train, cv=5, scoring="accuracy")
    print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
```

## Final model testing

```python
fmodel = LogisticRegression()
```

```python
X_test = df_test.drop(columns=["target", "target_name"]).values
Y_test = df_test["target"].values
```

```python
X_test.shape
```

```python
Y_test
```

```python
# Training final model using full train data
fmodel.fit(X_train, Y_train)
```

```python
y_test_pred = fmodel.predict(X_test)
```

```python
test_set_correctly_classified = y_test_pred == Y_test
test_set_accuracy = np.mean(test_set_correctly_classified)
```

```python
print(f"Test set accuracy: {test_set_accuracy * 100:.2f}%")
```

```python
df_predictions_test = df_test.copy()
df_predictions_test["correct_prediction"] = test_set_correctly_classified
df_predictions_test["prediction"] = y_test_pred
df_predictions_test["prediction_label"] = df_predictions_test["prediction"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
```

```python
df_predictions_test.head()
```

```python
plot_incorrect_predictions(df_predictions_test, "petal length (cm)", "petal width (cm)")
```
