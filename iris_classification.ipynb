# Creating a Pandas DataFrame using the data

```python
df = pd.DataFrame(data["data"], columns=data["feature_names"])
df["target"] = data["target"]
```

## Relationship of the data features with the target

```python
data["target_names"]
```

```python
# Creates new column in the dataframe using the species name
df["target_name"] = df["target"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
```

```python
col = "sepal length (cm)"
sns.relplot(x=col, y="target", hue="target_name", data=df)
_ = plt.suptitle(col, y=1.05)
```

```python
col = "sepal width (cm)"
sns.relplot(x=col, y="target", hue="target_name", data=df)
_ = plt.suptitle(col, y=1.05)
```

```python
col = "petal length (cm)"
sns.relplot(x=col, y="target", hue="target_name", data=df)
_ = plt.suptitle(col, y=1.05)
```

```python
col = "petal width (cm)"
sns.relplot(x=col, y="target", hue="target_name", data=df)
_ = plt.suptitle(col, y=1.05)
```
